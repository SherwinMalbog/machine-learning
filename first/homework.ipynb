```python
asia_df = df[df['origin'] == 'Asia'][['vehicle_weight', 'model_year']][:7]
```

```python
asia_df
```

```python
X = asia_df.values
```

```python
X
```

```python
XTX = X.T.dot(X)
XTX
```

```python
XTX_inv = np.linalg.inv(XTX)
XTX_inv
```

```python
y = np.array([1100, 1300, 800, 900, 1000, 1100, 1200])
```

```python
w = (XTX_inv @ X.T) @ y
w
```
